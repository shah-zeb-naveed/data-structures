```python
def binary_search(arr, low, high, x):
    if high >= low:
        mid = (low + high) // 2

        if arr[mid] == x:
            return mid

        elif arr[mid] < x:
            low = mid + 1
            return binary_search(arr, low, high, x)

        elif arr[mid] > x:
            high = mid - 1
            return binary_search(arr, low, high, x)

    else:
        return -1


def iterative_binary_search(arr, x):
    low = 0
    high = len(arr) - 1

    while low <= high:
        mid = (low + high) // 2

        if x == arr[mid]:
            return mid

        elif x > arr[mid]:
            low = mid + 1

        else:
            high = mid - 1

    return -1


def iterative_at_least_binary_search(arr, x):
    low = 0
    high = len(arr) - 1

    while low <= high:

        if high - low == 1:
            return low if x == arr[low] else high

        mid = (low + high) // 2

        if x == arr[mid]:
            return mid

        elif x > arr[mid]:
            low = mid  # have to retain

        else:
            high = mid  # have to retain

    return -1


def at_least_binary_search(arr, low, high, x):

    if high - low == 1:
        return low if arr[low] == low else high

    if high > low:

        mid = (low + high) // 2

        if arr[mid] == x:
            return mid

        elif arr[mid] < x:
            low = mid
            return at_least_binary_search(arr, low, high, x)

        elif arr[mid] > x:
            high = mid
            return at_least_binary_search(arr, low, high, x)

    else:
        return -1


arr = [2, 3, 5, 10, 40]
print(binary_search(arr, low=0, high=len(arr)-1, x=4))
print(iterative_binary_search(arr, x=4))
print(iterative_at_least_binary_search(arr, x=4))
print(at_least_binary_search(arr, low=0, high=len(arr)-1, x=4))
```

```python
# merge sort, divide and conquer

def merge_sort(arr):

    # base case
    if len(arr) <= 1:

        print("partial sort: ", arr)
        return arr

    left_arr = arr[:len(arr) // 2]
    right_arr = arr[len(arr) // 2:]

    # recursion. fixed by referece so no need to assign to new variable
    merge_sort(left_arr)
    merge_sort(right_arr)

    # merge
    left_idx = 0
    right_idx = 0
    merged_idx = 0

    while left_idx < len(left_arr) and right_idx < len(right_arr):
        if left_arr[left_idx] < right_arr[right_idx]:
            # fix arr in place but note that the arr passed could be a left or right array
            # itself. of courset the parent call's arr is the original arr.
            arr[merged_idx] = left_arr[left_idx]
            left_idx += 1

        else:
            arr[merged_idx] = right_arr[right_idx]
            right_idx += 1

        merged_idx += 1

    while right_idx < len(right_arr):
        arr[merged_idx] = right_arr[right_idx]
        right_idx += 1
        merged_idx += 1

    while left_idx < len(left_arr):
        arr[merged_idx] = left_arr[left_idx]
        left_idx += 1
        merged_idx += 1

    print("partial sort: ", arr)

    return arr

arr_test = [4, 3, 2, 1, 8]
arr_test = merge_sort(arr_test)

print(arr_test)
```

```python
# Python program for implementation of Selection Sort

def selection_sort(A):
    # Traverse through all array elements
    for i in range(len(A)):

        # Find the minimum element in remaining
        # unsorted array
        min_idx = i
        for j in range(i+1, len(A)):
            if A[min_idx] > A[j]:
                min_idx = j

        # Swap the found minimum element with
        # the first element
        A[i], A[min_idx] = A[min_idx], A[i]

A = [64, 25, 12, 22, 11]
selection_sort(A)
print(A)
```

```python
# Counting Sort

# The main function that sort the given string arr[] in
# alphabetical order

def countSort(arr):

    output = [0] * len(arr)
    count = [0] * 256

    # For storing the resulting answer since the
    # string is immutable
    ans = [""] * len(arr)

    for i in arr:
        count[ord(i)] += 1

    # Change count[i] so that count[i] now contains actual
    # position of this character in output array
    for i in range(1, 256):
        count[i] += count[i-1]

    for i in range(len(arr)):
        # note: a count of an element present in input data can never be 0 for obvious reasons but the index starts with 0 so we have to subtract 1 to get index.
        output[count[ord(arr[i])]-1] = arr[i]

        # also, we first insert an element at it's highest place and if we have more of the same element coming in, we keep on place it one index earlier
        # this essentially has the same effect as right shifting the cum-sum array and keep on insert one element after instead of earlier.
        count[ord(arr[i])] -= 1

    ans[:] = output[:]
        
    return ans

arr = "geeksforgeeks"
ans = countSort(arr)
print("".join(ans))

# This code is contributed by Nikhil Kumar Singh
```

```python
# Radix Sort according to the digit represented by exp.

def countingSort(arr, exp1):

    """
    
    this is a variant of count sort but the core algorithm is still the same.
    it just performs the same operations using the exp1 provided.

    """

    n = len(arr)
    output = [0] * (n)
    count = [0] * (10)

    for i in range(n):
        digit = (arr[i] // exp1) % 10
        # mod means 2, 22, 222, 232, 242, 11111112, all will be counted in this. we are just preparing counts of last digit here
        count[digit] += 1

    # cum sum
    for i in range(1, 10):
        count[i] += count[i - 1]

    # should start from end as the arr is already sorted
    # based on previous LSBs or whatever.
    # since we first fill in highest position first and then
    # decremenet count -= 1, we should first place whatever is
    # the highest and that will be in the reverse order.
    i = n - 1
    while i >= 0:
        digit = (arr[i] // exp1) % 10
        output[count[digit] - 1] = arr[i]
        count[digit] -= 1
        i += 1

    arr[:] = output[:]


def radixSort(arr):

    # Find the maximum number to know number of digits
    max1 = max(arr)

    # Do counting sort for every digit. Note that instead
    # of passing digit number, exp is passed. exp is 10^i
    # where i is current digit number
    exp = 1
    while max1 / exp > 1:  # loops number of digits
        print(arr)
        countingSort(arr, exp)
        exp *= 10


arr = [9, 1, 2, 5, 3, 11, 6, 2]
radixSort(arr)
print(arr)
# This code is contributed by Mohit Kumra
# Edited by Patrick Gallagher
```

```python
# heap Sort

def heapifyDown(arr, n, i):
    largest = i  # Initialize largest as root
    l = 2 * i + 1	 # left = 2*i + 1
    r = 2 * i + 2	 # right = 2*i + 2

    # See if left child of root exists and is
    # greater than root
    if l < n and arr[largest] < arr[l]:
        largest = l

    # See if right child of root exists and is
    # greater than root
    if r < n and arr[largest] < arr[r]:
        largest = r

    # Change root, if needed
    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]  # swap

        # Heapify the area that got potentially disturbed.
        heapifyDown(arr, n, largest)

def heapSort(arr):
    n = len(arr)
    # Build a maxheap.

    # In the case of a complete tree, the first (read: last?) index of a non-leaf node is given by n/2 - 1.
    # All other nodes after that are leaf-nodes and thus don't need to be heapified.
    # starting from bottom is the key here for O(n)



    # Equivalent to Heapify in O(n)
    for i in range(n//2 - 1, -1, -1):
        heapifyDown(arr, n, i)
        print(arr)

    print('------------------------')
    # Once heapified, one by one extract elements
    for i in range(n-1, 0, -1):  # ascending true
        arr[i], arr[0] = arr[0], arr[i]  # swap
        print(arr)

        # Although the tree is shrinking, it doesn't shrink fast enough: The height of the tree stays constant until
        # you have removed the first half of the nodes (when you clear out the bottom layer completely).

        heapifyDown(arr, i, 0) # we are essentially decreasing the size of the heap
        print(arr)

arr = [3, 0, 11, 5, 6, 7]
heapSort(arr)
print(arr)
```
