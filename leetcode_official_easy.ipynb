```python
class Solution:
    def isValidSudoku(self, board: List[List[str]]) -> bool:
        N = 9
        # Use binary number to check previous occurrence
        rows = [0] * N
        cols = [0] * N
        boxes = {}

        for r in range(N):
            for c in range(N):
                # Check if the position is filled with number
                if board[r][c] == ".":
                    continue

                pos = int(board[r][c]) - 1 # elements start from 1 and bit positions start from 0 obviously

                # Check the row if this element (with associated position) is already 1
                if rows[r] & (1 << pos):
                    return False
                rows[r] |= (1 << pos) # light it up

                # Check the column
                if cols[c] & (1 << pos):
                    return False
                cols[c] |= (1 << pos)

                # Check the box
                #idx = (r // 3) * 3 + c // 3
                # just go with easier stuff
                idx = (r // 3, c // 3)
                
                if not idx in boxes:
                    boxes[idx] = 0
                
                if boxes[idx] & (1 << pos):
                    return False
                boxes[idx] |= (1 << pos)

        return True
    

class Solution2:
    def isValidSudoku(self, board: List[List[str]]) -> bool:
        
        rows = len(board)
        cols = len(board[0])
        
        for row in range(rows):
            
            elements = set()
            
            for col in range(cols):  
                x = board[row][col]
                
                if x in elements and x != '.':
                    print('Row: ', row)
                    print('Col: ', col)
                    print('Failure 1')
                    return False
                elements.add(x)
                    
        for col in range(cols):
            
            elements = set()
            
            for row in range(rows):
                x = board[row][col]
                
                if x in elements and x != '.':
                    print('Failure 2')
                    return False
                
                elements.add(x)
        
        for v in range(0, 9, 3):
            for h in range(0, 9, 3):
                
                elements = set()
                
                for row in range(3):
                    for col in range(3):
                        x = board[row+v][col+h]
                        
                        if x in elements and x != '.':
                            print('Failure 3')
                            return False 
                        elements.add(x)
                            
        # here although this is O(N), this is actually O(3N)
        # benefit is that in big o terms, it is still just N 
        # and it does not have to hold more memeory than N
        
        # alternatively each box can be mapped using (row//3, col//3) tuple
        # {(row//3, col//3):set()} 
        # this will reduce O(3N) to O(N) practically but end up using more space
                            
        return True
                
                
                
                
                
                
            
                
                    
            
```

```python
class Solution:
    def reverse(self, x: int) -> int:
        
        negative = False
        
        if x < 0:
            negative = True
            x = x * -1
            
        final = 0
        
        while x > 0:
            digit = x % 10
            x = x // 10
            final = final * 10 + digit
            
        if negative:
            final *= -1
            if final < -1 * 2**(31):
                return 0
        else:
            if final > 2**(31) - 1:
                return 0
            
        return final
```

```python
class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        
        if len(s) != len(t):
            return False
        
        freqs = [0] * 26
        
        for c in s:
            freqs[ord(c) - ord('a')] += 1
        
        for c in t:
            freqs[ord(c) - ord('a')] -= 1
            
            if freqs[ord(c) - ord('a')] < 0:
                return False
            
        for freq in freqs:
            if freq > 0:
                return False
            
        return True
        
        
"""

Follow up

What if the inputs contain unicode characters? How would you adapt your solution to such case?

Answer

Use a hash table instead of a fixed size counter. Imagine allocating a large size array to fit the entire range of unicode characters, which could go up to more than 1 million. A hash table is a more generic solution and could adapt to any range of characters.

"""
```

```python
class Solution:
    def fizzBuzz(self, n: int) -> List[str]:

        # In Approach 3, more computation (divisions) are performed. We divide every number by 3 & 5. Instead, if-else ladder like if(n%15==0) elseif(n%5==0) elseif(n%3==0) will end up in much less divisions.
        
        # Approach 3 is for the most generalized case where there are, for instance, a 1000 numbers to check for. Building an if else ladder is infeasible and a more generalized solution is needed. We make the trade off for achieving simplicity and more maintainable/extensible code by making those additional modulus checks.
        
        def get_mapping(i):
            
            s = ''
            
            if i % 3 == 0:
                s += 'Fizz'
            
            if i % 5 == 0:
                s += 'Buzz'
            
            if len(s) == 0:
                return str(i)
            
            return s
        
        def get_mapping_2(i):
            
            mapper = {3 : "Fizz", 5 : "Buzz"}
            
            s = ''
            
            for key in mapper:
                
                if i % key == 0:
                    s += mapper[key]
                        
            if len(s) == 0:
                return str(i)
            
            return s
        
        answer = [get_mapping_2(i) for i in range(1, n+1)]
        
        return answer
```

```python
class Solution:
    
    def countPrimes(self, n: int) -> int:
        
        if n <= 2:
            return 0
        
        numbers = [False, False] + [True] * (n - 2)
        
        # claim # 1: for sqrt(n) + 1 and onwards, all the composites
        # until n will be covered by the primes below sqrt(n).
        # The Sieve of Eratosthenes code on Wikipedia is intended to generate a list of all primes up to n
        # Suppose k≤n is not prime, so we have two factors a,b. We can't have a,b both larger than n−−√, as
        # then k=ab would be larger than n. Thus in order to show that k is prime, we only need to check that
        # it is not divisible by a number up to n−−√. Equivalently (and this is the key insight in the Sieve)
        # we only need to check that it is not divisible by a prime up to n−−√.
        
        for p in range(2, int(sqrt(n)) + 1):
          
            if numbers[p]:
                
                # claim # 2: for a prime number p, smaller prime numbers
                # will have marked composites until p * p.
                # 2*5 already crossed 10 so 5*2 not needed. 3*5 already crossed 15 so 5*3 not needed...
                
                for multiple in range(p * p, n, p):
                    numbers[multiple] = False
                    
        return sum(numbers)
```

```python
class Solution:
    def isValid(self, s: str) -> bool:
        
        stack = []
        
        matching = {
            "(": ")",
            "[": "]",
            "{": "}"
        }

        for bracket in s:
            if bracket in matching:
                stack.append(bracket)
            else:
                if len(stack) == 0:
                    return False

                popped = stack.pop()

                if matching[popped] == bracket:
                    pass
                else:
                    return False

        return len(stack) == 0
            
```

```python
class Solution:
    
    """
    
    Note the difference between reverse and inverse the bits.
    
    """
    
    def reverseBits(self, n: int) -> int:
        
        
        # solution 1:
        
        rev = 0
        power = 31 # max 0-index based on 32-bit integer condition
        
        while n:
            extracted_last_bit = (n & 1) # or % 2
            extracted_bit_at_the_right_place =  extracted_last_bit << power
            # add to result
            rev = rev | extracted_bit_at_the_right_place # or simple +
            # not sure why rev = rev + (n & 1) << power gives such a weird answer
            
            # discard the processed bit
            n = n >> 1
            power -= 1
            
        return rev
        
        # sol: return int(bin(n)[:1:-1], 2)
        
        
        
```

```python
class MinStack:

    def __init__(self):
        self.stack = None

    def push(self, val: int) -> None:
        
        if not self.stack:
            self.stack = [(val, val)]
            return
        
        cur_min = self.getMin()
        
        if val < cur_min:
            cur_min = val
        
        self.stack.append((val, cur_min))
        
    def pop(self) -> None:
        self.stack.pop()

    def top(self) -> int:
        return self.stack[-1][0]
        
    def getMin(self) -> int:        
        return self.stack[-1][1]     


# Your MinStack object will be instantiated and called as such:
# obj = MinStack()
# obj.push(val)
# obj.pop()
# param_3 = obj.top()
# param_4 = obj.getMin()
```
