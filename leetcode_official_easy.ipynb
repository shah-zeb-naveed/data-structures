```python
class Solution:
    def isValidSudoku(self, board: List[List[str]]) -> bool:
        
        rows = len(board)
        cols = len(board[0])
        
        for row in range(rows):
            
            elements = set()
            
            for col in range(cols):  
                x = board[row][col]
                
                if x in elements and x != '.':
                    print('Row: ', row)
                    print('Col: ', col)
                    print('Failure 1')
                    return False
                else:
                    elements.add(x)
                    
        for col in range(cols):
            
            elements = set()
            
            for row in range(rows):
                x = board[row][col]
                
                if x in elements and x != '.':
                    print('Failure 2')
                    return False
                
                else:
                    elements.add(x)
        
        for v in range(0, 9, 3):
            for h in range(0, 9, 3):
                
                visited = set()
                
                for row in range(3):
                    for col in range(3):
                        x = board[row+v][col+h]
                        
                        if x in visited and x != '.':
                            print('Failure 3')
                            return False 
                        else:
                            visited.add(x)
                            
        return True
                            
# here although this is O(N), this is actually O(3N)
# benefit is that in big o terms, it is still just N 
# and it does not have to hold more memeory than N

# alternatively each box can be mapped using (row//3, col//3) tuple
# {(row//3, col//3):set()} 
# this will reduce O(3N) to O(N) practically but end up using more space           
```
