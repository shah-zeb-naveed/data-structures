```python
def find_position(x, arr):
    if len(arr) == 1:
        if x <= arr[0]:
            return 1

        elif x > arr[0]:
            return -1

    else:
        middle = len(arr) // 2
        if x == arr[middle]:
            return middle

        elif x > arr[middle]:
            domain = arr[middle+1:]

        else:
            domain = arr[:middle]

        return middle + find_position(x, domain)
```
