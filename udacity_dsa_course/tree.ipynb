```python
class Node(object):
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

class BinaryTree(object):
    def __init__(self, root):
        self.root = Node(root)

    def search(self, find_val):
        """Return True if the value
        is in the tree, return
        False otherwise."""
        
        # have to use a helper function to avoid this statement
        curr = self.root

        if self.preorder_search(curr, find_val):
            return True
        
        return False

    def print_tree(self):
        """Print out all tree nodes
        as they are visited in
        a pre-order traversal."""
        
        start = self.root
        traversal = []
        self.preorder_print(start, traversal)
        
        return '-'.join(map(str, traversal))
        

    def preorder_search(self, start, find_val):
        """Helper method - use this to create a 
        recursive search solution."""
        
        if not start:
            return False
            
        if start.value == find_val:
            return True
            
        if self.preorder_search(start.right, find_val):
            return True
            
        if self.preorder_search(start.left, find_val):
            return True
            
    def preorder_print(self, start, traversal):
        """Helper method - use this to create a 
        recursive print solution."""
        
        if start:
            traversal.append(start.value)
            self.preorder_print(start.left, traversal)
            self.preorder_print(start.right, traversal)

# Set up tree
tree = BinaryTree(1)
tree.root.left = Node(2)
tree.root.right = Node(3)
tree.root.left.left = Node(4)
tree.root.left.right = Node(5)

# Test search
# Should be True
print tree.search(4)
# Should be False
print tree.search(6)

# Test print_tree
# Should be 1-2-4-5-3
print tree.print_tree()
```

```python
class Node(object):
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

class BST(object):
    
    def __init__(self, root):
        self.root = Node(root)

    def insert(self, new_val):
        
        if not self.root:
            # won't even run since root needed for object init.
            print("Defining root")
            self.root = Node(new_val)
            
        curr = self.root
        
        while curr:
            
            if new_val < curr:
                if curr.left:
                    curr = curr.left
                else:
                    curr.left = Node(new_val)
                    break
        
            elif new_val > curr:
                if curr.right:
                    curr = curr.right
                else:
                    curr.right = Node(new_val)
                    break
            

    def search(self, find_val):
        
        if not self.root:
            return False
            
        curr = self.root
        
        while curr:
            
            if curr.value == find_val:
                return True
            
            elif curr.value < find_val:
                curr = curr.right
            
            else:
                curr = curr.left
        
        return False
    
# Set up tree
tree = BST(4)

# Insert elements
tree.insert(2)
# tree.insert(1)
# tree.insert(3)
# tree.insert(5)

# Check search
# Should be True
print tree.search(4)
# Should be False
print tree.search(6)
```
