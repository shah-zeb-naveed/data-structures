```python
class Solution:
    def productExceptSelf(self, nums: List[int]) -> List[int]:
        
        left = []
        
        for i in range(len(nums)):
            if i == 0:
                left.append(1)
                
            else:
                left.append(nums[i-1] * left[i-1])

        # space optimization by maintaining a single variable on the 2nd pass
        right = 1
    
        for i in reversed(range(len(nums))):
            left[i] = left[i] * right
            right *= nums[i]
            
        return left
```

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class BSTIterator:

    # When analyzing amortized time complexities, I find it easiest to reason that each node gets pushed and popped exactly once in next() when # 
    # iterating over all N nodes.
    # That comes out to 2N * O(1) over N calls to next(), making it O(1) on average, or O(1) amortized.

    def __init__(self, root: Optional[TreeNode]):
        
        self.stack = []
        
        # init minimum non-existent node
        node = root
        
        while node:
            self.stack.append(node)
            node = node.left            

    def next(self) -> int:
        
        popped = self.stack.pop()
        
        node = popped.right
                    
        while node:    
            self.stack.append(node)
            node = node.left
            
        return popped.val

    def hasNext(self) -> bool:
        
        if self.stack:
            return True
        
        else:
            return False
        


# Your BSTIterator object will be instantiated and called as such:
# obj = BSTIterator(root)
# param_1 = obj.next()
# param_2 = obj.hasNext()
```

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    
    diameter = float('-inf')
    
    def get_height_update_diameter(self, root):
                
        if not root:
            return 0
        
        left_height = self.get_height_update_diameter(root.left)
        right_height = self.get_height_update_diameter(root.right)
        
        dia = left_height + right_height
        
        if self.diameter < dia:
            self.diameter = dia
        
        return 1 + max(left_height, right_height)
        
    def diameterOfBinaryTree(self, root: Optional[TreeNode]) -> int:
        
        self.get_height_update_diameter(root)
        
        return self.diameter
```

```python
from collections import defaultdict

class DAG:

    def __init__(self, v=None):
        self.adj = defaultdict(list)
        self.v = v
    
    def add_edge(self, parent, dependent):
        self.adj[parent].append(dependent)

    def topological_sort_helper(self, c, output, visited, unique):

        visited.add(c)
        
        # for defaultdict, we don't need this check actually
        if c in self.adj:
            for dependent in self.adj[c]:
                if dependent not in visited and dependent in unique:
                    self.topological_sort_helper(dependent, output, visited, unique)

        output.append(c)

    def topological_sort(self, unique):
        output = []
        visited = set()

        # looping through all possile vertices and not just sources coz need to add to stack
        for i in range(self.v):
            
            # we can't infer ordering for characters other than the ones in unique set
            if len(visited) == len(unique):
                break
            
            # convert index to char
            c = chr(i + ord('a'))
            
            if c not in visited and c in unique:
                self.topological_sort_helper(c, output, visited, unique)

        return output
```

```python
from collections import defaultdict

class Graph:

    def __init__(self, v):
        self.adj = defaultdict(list)
        self.v = v
    
    def add_edge(self, u, v):
        self.adj[u].append(v)
        self.adj[v].append(u)

    def is_bipartite_helper(self, queue, colors):
        while queue:
            x = queue.pop(0) # should've used an actual queue but anyways...

            for neib in self.adj[x]:
                if neib == x:
                    # self loop
                    return False

                if colors[neib] == 0:
                    colors[neib] = colors[x] * -1 # invert colors
                    queue.append(neib)

                elif colors[neib] ==  colors[x]:
                    return False
        return True

    def is_bipartite_disconnected(self):
        colors = [0] * self.v
        queue = [0]
        colors[0] = 1

        # i think this is only needed for disconnected graphs
        for i in range(self.v):
            if colors[i] != 0:
                if not self.is_bipartite_helper(queue, colors):
                    return False

        return True

    def is_bipartite(self):
        colors = [0] * self.v
        queue = [0]
        colors[0] = 1

        while queue:
            x = queue.pop(0) # should've used an actual queue but anyways...

            # check/set colors of all neighbours
            for neib in self.adj[x]:
                if neib == x:
                    # self loop
                    return False

                if colors[neib] == 0:
                    colors[neib] = colors[x] * -1 # invert colors
                    queue.append(neib)

                elif colors[neib] ==  colors[x]:
                    return False
        return True

    def is_bipartite_dfs(self):
        
        def helper(v):
            for neib in self.adj[v]:
                if colors[neib] == colors[v]:
                    return False

                # if not visited
                elif colors[neib] == 0:                
                    colors[neib] = colors[v] * - 1
                    if not helper(neib):
                        return False
            return True

        colors = [0] * self.v

        for i in range(self.v):
            if colors[i] == 0:
                colors[i] = 1
                if not helper(i):
                    return False

        return True


dag = Graph(5)
dag.add_edge(0, 1)
dag.add_edge(0, 2)
dag.add_edge(2, 3)
dag.add_edge(3, 1)
print(dag.is_bipartite())
print(dag.is_bipartite_dfs())

dag = Graph(5)
dag.add_edge(0, 1)
dag.add_edge(0, 2)
dag.add_edge(2, 4)
dag.add_edge(3, 1)
dag.add_edge(3, 4)
print(dag.is_bipartite())
print(dag.is_bipartite_dfs())

# dag = Graph(5)
# dag.add_edge(0, 1)
# dag.add_edge(0, 2)
# dag.add_edge(2, 4)
# dag.add_edge(3, 1)
# dag.add_edge(3, 4)
# print(dag.is_bipartite())
```

```python
"""
# Definition for a Node.
class Node:
    def __init__(self, x: int, next: 'Node' = None, random: 'Node' = None):
        self.val = int(x)
        self.next = next
        self.random = random
"""

class Solution:

    def __init__(self):
        self.hashmap = {}
    
    # this is beautiful
    def copyRandomList_dfs(self, head):

        if head == None:
            return None

        if head in self.hashmap:
            return self.hashmap[head]

        node = Node(head.val)
        self.hashmap[head] = node

        # Recursively copy the remaining linked list starting once from the next pointer and then from the random pointer.
        # Thus we have two independent recursive calls.
        # Finally we update the next and random pointers for the new node created.
        node.next = self.copyRandomList(head.next)
        node.random = self.copyRandomList(head.random)

        return node
    
    def copyRandomList2(self, head: 'Optional[Node]') -> 'Optional[Node]':
        
        node = head
        
        # dummy
        dummy = Node(0)
        prev = dummy
        
        hashmap = {}
        
        while node:
            if node not in hashmap:
                node_copy = Node(node.val)
                hashmap[node] = node_copy
            
            node_copy = hashmap[node]
            prev.next = node_copy
            
            # random pointer
            if node.random:
                if node.random not in hashmap:
                    hashmap[node.random] = Node(node.random.val)

                node_copy.random = hashmap[node.random]    
            else:
                node_copy.random = None
                
            prev = node_copy
            node = node.next
        
        prev.next = None
        head = dummy.next
        
        return head
    
    # weaving of linked lists but modifies the parameter
    def copyRandomList(self, head):
        """
        :type head: Node
        :rtype: Node
        """
        if not head:
            return head

        # Creating a new weaved list of original and copied nodes.
        ptr = head
        while ptr:

            # Cloned node
            new_node = Node(ptr.val, None, None)

            # Inserting the cloned node just next to the original node.
            # If A->B->C is the original linked list,
            # Linked list after weaving cloned nodes would be A->A'->B->B'->C->C'
            new_node.next = ptr.next
            ptr.next = new_node
            
            ptr = ptr.next.next

        ptr = head

        # Now link the random pointers of the new nodes created.
        # Iterate the newly created list and use the original nodes random pointers,
        # to assign references to random pointers for cloned nodes.
        while ptr:
            ptr.next.random = ptr.random.next if ptr.random else None
            
            ptr = ptr.next.next

        # Unweave the linked list to get back the original linked list and the cloned list.
        # i.e. A->A'->B->B'->C->C' would be broken to A->B->C and A'->B'->C'
        head_new = head.next
        
        ptr_old_list = head # A->B->C
        ptr_new_list = head.next # A'->B'->C'
        
        while ptr_old_list:
            
            # fix original to point to its next original
            # no need to worry about None.next issue as next will always be clone
            # at the very end, it will point to None of course
            ptr_old_list.next = ptr_old_list.next.next
            
            # fix clone to point to its next clone
            # as interleaved list will always end on a clone, we need to check if its next is None or not
            ptr_new_list.next = ptr_new_list.next.next if ptr_new_list.next else None
            
            ptr_old_list = ptr_old_list.next
            ptr_new_list = ptr_new_list.next
            
        return head_new
    
    def copyRandomList_1(self, head: 'Optional[Node]') -> 'Optional[Node]':
        
        # doesnt work on duplicates
        node = head
        
        # dummy
        dummy = Node(0)
        prev = dummy
        
        hashmap = {}
        
        while node:
            
            if node.val not in hashmap:
                node_copy = Node(node.val)
                hashmap[node.val] = node_copy
            
            node_copy = hashmap[node.val]
            prev.next = node_copy
            
            # random pointer
            if node.random:
                rand_val = node.random.val
                if rand_val not in hashmap:
                    nxt_rand = Node(rand_val)
                    hashmap[rand_val] = nxt_rand

                node_copy.random = hashmap[rand_val]    
            else:
                node_copy.random = None
                
            prev = node_copy
            node = node.next
        
        prev.next = None
        head = dummy.next
        
        return head
```

```python
class Solution:
    def merge(self, intervals: List[List[int]]) -> List[List[int]]:
        
        intervals.sort()
        
        ans = [intervals[0]]
        
        for i in range(1, len(intervals)):
            
            start, end = intervals[i]
            cur_start, cur_end = ans[-1]
            
            if start <= cur_end:
                ans[-1] = [cur_start, max(end, cur_end)]
                    
            else:
                ans.append(intervals[i])
                
        return ans
        

    # Connected Components Approach
    def overlap(self, a, b):
        return a[0] <= b[1] and b[0] <= a[1]

    # generate graph where there is an undirected edge between intervals u
    # and v iff u and v overlap.
    def buildGraph(self, intervals):
        graph = collections.defaultdict(list)
 
        for i, interval_i in enumerate(intervals):
            for j in range(i+1, len(intervals)):
                if self.overlap(interval_i, intervals[j]):
                    graph[tuple(interval_i)].append(intervals[j])
                    graph[tuple(intervals[j])].append(interval_i)

        return graph

    # merges all of the nodes in this connected component into one interval.
    def mergeNodes(self, nodes):
        min_start = min(node[0] for node in nodes)
        max_end = max(node[1] for node in nodes)
        return [min_start, max_end]

    # gets the connected components of the interval overlap graph.
    def getComponents(self, graph, intervals):
        visited = set()
        comp_number = 0
        nodes_in_comp = collections.defaultdict(list)

        def markComponentDFS(start):
            stack = [start]
            while stack:
                node = tuple(stack.pop())
                if node not in visited:
                    visited.add(node)
                    nodes_in_comp[comp_number].append(node)
                    stack.extend(graph[node])

        # mark all nodes in the same connected component with the same integer.
        for interval in intervals:
            if tuple(interval) not in visited:
                markComponentDFS(interval)
                comp_number += 1

        return nodes_in_comp, comp_number


    def merge1(self, intervals: List[List[int]]) -> List[List[int]]:
        graph = self.buildGraph(intervals)
        nodes_in_comp, number_of_comps = self.getComponents(graph, intervals)

        # all intervals in each connected component must be merged.
        return [self.mergeNodes(nodes_in_comp[comp]) for comp in range(number_of_comps)]
```

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right

from collections import defaultdict
from queue import Queue

class Solution:
    
    def verticalOrder(self, root: Optional[TreeNode]) -> List[List[int]]:
        
        if not root:
            return 
        
        q = Queue()
        q.put((root, 0))
        res = defaultdict(list)
        min_column = 0
        max_column = 0
        
        while not q.empty():
            
            node, col = q.get()
            
            res[col].append(node.val)
            min_column = min(min_column, col)
            max_column = max(max_column, col)

            if node.left:
                q.put((node.left, col - 1))
            
            if node.right:
                q.put((node.right, col + 1))
                
        return [res[col] for col in range(min_column, max_column + 1)]
    

    def verticalOrder_DFS(self, root: TreeNode) -> List[List[int]]:
        if not root:
            return []

        columnTable = defaultdict(list)
        min_column = max_column = 0

        # it is clearly a top to bottom (implying bfs)
        # to my surprise, there is actually a way via dfs as well.
        # this additinally requires sorting by row as well which
        # comes natural in bfs
        
        def DFS(node, row, column):
            if node is not None:
                
                nonlocal min_column, max_column
                
                columnTable[column].append((row, node.val))
                
                min_column = min(min_column, column)
                max_column = max(max_column, column)

                DFS(node.left, row + 1, column - 1)
                DFS(node.right, row + 1, column + 1)

        DFS(root, 0, 0)

        # order by column and sort by row
        ret = []
        for col in range(min_column, max_column + 1):
            columnTable[col].sort(key=lambda x:x[0])
            colVals = [val for row, val in columnTable[col]]
            ret.append(colVals)

        return ret
```

```python
class Solution:
    def getSum(self, a: int, b: int) -> int:
        x, y = abs(a), abs(b)
        # ensure x >= y
        if x < y:
            return self.getSum(b, a)  
        sign = 1 if a > 0 else -1
        
        if a * b >= 0:
            # sum of two positive integers
            while y:
                x, y = x ^ y, (x & y) << 1
        else:
            # difference of two positive integers
            while y:
                x, y = x ^ y, ((~x) & y) << 1
        
        return x * sign
```

```python
from queue import Queue

class MovingAverage:

    def __init__(self, size: int):
        self.q = Queue()
        self.last_sum = 0
        self.size = size

    def next(self, val: int) -> float:
        
        self.q.put(val)
        size = self.q.qsize()
        
        if size <= self.size:
            self.last_sum = self.last_sum + val
            return self.last_sum / size
    
        self.last_sum = self.last_sum - self.q.get() + val
        return self.last_sum / self.size
            

# Your MovingAverage object will be instantiated and called as such:
# obj = MovingAverage(size)
# param_1 = obj.next(val)
```

```python
class Solution:
    
    def shipWithinDays(self, weights: List[int], D: int) -> int:
        
        low, high = max(weights), sum(weights)
        
        while low < high:
            capacity_mid = (low + high) // 2
            days = self.findDays(weights, capacity_mid)
            
            if days <= D: # maybe with same days we still have a better solution
                high = capacity_mid
            
            elif days > D:
                low = capacity_mid + 1
        
        return low
        
        
    def findDays(self, weights, capacity):
        days = 1
        cur_w = 0
        
        for w in weights:
            
            # need another day
            if w + cur_w > capacity:
                cur_w = w
                days += 1
                
            # can manage in the same day
            else:
                cur_w += w
                
        return days
```

```python
class Solution:
    def removeInvalidParentheses(self, s):
        """
        :type s: str
        :rtype: List[str]
        """

        left = 0
        right = 0

        # First, we find out the number of misplaced left and right parentheses.
        for char in s:
            if char == '(':
                left += 1
                
            elif char == ')':
                right = right + 1 if left == 0 else right
                left = left - 1 if left > 0 else left
        
        result = set()
        
        def recurse(s, index, left_count, right_count, left_rem, right_rem, expr):

            if index == len(s):
                #  if we have removed the total number of left and right
                # parentheses that we should have removed.
                if left_rem == 0 and right_rem == 0:
                    ans = "".join(expr)
                    result.add(ans)
                    
                return
            
            c = s[index]
            
            # The discard case. Note that here we have our pruning condition.
            # We don't recurse if the remaining count for that parenthesis is == 0.
            if (c == '(' and left_rem > 0) or (c == ')' and right_rem > 0):
                recurse(s,
                        index + 1,
                        left_count,
                        right_count,
                        left_rem - (c == '('),
                        right_rem - (c == ')'),
                        expr)

            expr.append(c)    

            # Simply recurse one step further if the current character is not a parenthesis.
            if not c in ['(', ')']:
                recurse(s,
                        index + 1,
                        left_count,
                        right_count,
                        left_rem,
                        right_rem,
                        expr)

            elif c == '(':
                recurse(s,
                        index + 1,
                        left_count + 1,
                        right_count,
                        left_rem,
                        right_rem,
                        expr)

            elif s[index] == ')' and left_count > right_count:
                recurse(s,
                        index + 1,
                        left_count,
                        right_count + 1,
                        left_rem,
                        right_rem,
                        expr)

            # backtrack
            expr.pop()                 

        # Now, the left and right variables tell us the number of misplaced left and
        # right parentheses and that greatly helps pruning the recursion.
        recurse(s, 0, 0, 0, left, right, [])     
        return list(result)
    
    
class Solution1(object):

    def __init__(self):
        self.valid_expressions = None
        self.min_removed = None

    def reset(self):
        self.valid_expressions = set()
        self.min_removed = float("inf")

    """
        string: The original string we are recursing on.
        i: current index in the original string.
        left: number of left parentheses till now.
        right: number of right parentheses till now.
        ans: the resulting expression in this particular recursion.
        rem_count: number of parentheses ignored in this particular recursion.
    """
    def remaining(self, string, i, left_count, right_count, expr, rem_count):

        if i == len(string):

            # If the current expression is valid. The only scenario where it can be
            # invalid here is if left > right. The other way around we handled early on in the recursion.
            if left_count == right_count:

                if rem_count <= self.min_removed:
                    # This is the resulting expression.
                    # Strings are immutable in Python so we move around a list in the recursion
                    # and eventually join to get the final string.
                    possible_ans = "".join(expr)

                    # If the current count of brackets removed < current minimum, ignore
                    # previous answers and update the current minimum count.
                    if rem_count < self.min_removed:
                        self.valid_expressions = set()
                        self.min_removed = rem_count

                    self.valid_expressions.add(possible_ans)    
                    
            return

        c = string[i]
        print(expr, c)      

        # If the current character is not a parenthesis, just recurse one step ahead.
        if not c in ['(', ')']:
            expr.append(c)
            self.remaining(string, i + 1, left_count, right_count, expr, rem_count)
            
            # backtrack since we wanna go back and and clean up. need to recurse without this non-bracket character
            expr.pop()
            return

        # Also,        
        expr.append(c)

        # If the current parenthesis is an opening bracket, we consider it
        # and increment left and  move forward
        if c == '(':
            self.remaining(string, i + 1, left_count + 1, right_count, expr, rem_count)
        
        elif right_count < left_count: # optimization. # note that right_count here is still coming from the parent call.
            # If the current parenthesis is a closing bracket, we consider it only if we
            # have more number of opening brackets and increment right and move forward.
            self.remaining(string, i + 1, left_count, right_count + 1, expr, rem_count)
        
        # backtrack
        expr.pop()
        
        # One recursion is with ignoring the current character.
        # So, we increment the ignored counter and leave the left and right untouched.
        self.remaining(string, i + 1, left_count, right_count, expr, rem_count + 1)

    def removeInvalidParentheses(self, s):
        """
        :type s: str
        :rtype: List[str]
        """

        # Reset the class level variables that we use for every test case.
        self.reset()

        # Recursive call
        self.remaining(s, 0, 0, 0, [], 0)
        return list(self.valid_expressions)
```

```python
class Solution(object):
    def largestIsland(self, grid):
        N = len(grid)

        def neighbors(r, c):
            for nr, nc in ((r-1, c), (r+1, c), (r, c-1), (r, c+1)):
                if 0 <= nr < N and 0 <= nc < N:
                    yield nr, nc

        def dfs(r, c, index):
            ans = 1
            grid[r][c] = index # alternative for visited. 2-in-1 as it groups by.
            for nr, nc in neighbors(r, c):
                if grid[nr][nc] == 1:
                    ans += dfs(nr, nc, index)
            return ans

        area = {}
        index = 2
        
        for r in range(N):
            for c in range(N):
                if grid[r][c] == 1:
                    area[index] = dfs(r, c, index)
                    index += 1

        ans = max(area.values() or [0])
        
        for r in range(N):
            for c in range(N):
                if grid[r][c] == 0:
                    seen = {grid[nr][nc] for nr, nc in neighbors(r, c) if grid[nr][nc] > 1}
                    ans = max(ans, 1 + sum(area[i] for i in seen))
        return ans
    
    
class Solution1(object):
    
    def largestIsland(self, grid):
        N = len(grid)

        def dfs(r, c):
            
            seen = {(r, c)}
            stack = [(r, c)]
            
            while stack:
                r, c = stack.pop()
                for nr, nc in [(r-1, c), (r, c-1), (r+1, c), (r, c+1)]:
                    if (nr, nc) not in seen and 0 <= nr < N and 0 <= nc < N and grid[nr][nc]:
                        stack.append((nr, nc))
                        seen.add((nr, nc))
            
            return len(seen)

        ans = 0
        has_zero = False
        for r, row in enumerate(grid):
            for c, val in enumerate(row):
                if val == 0:
                    has_zero = True
                    grid[r][c] = 1
                    ans = max(ans, dfs(r, c))
                    grid[r][c] = 0

        return ans if has_zero else N*N
```
