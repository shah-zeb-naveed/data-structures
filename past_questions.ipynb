```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def rangeSumBST(self, root: Optional[TreeNode], low: int, high: int) -> int:
        
        def traverse(node, low, high):
            
            if not node:
                return 0
            
            subtree_sum = 0
            
            if node.val >= low and node.val <= high:
                subtree_sum += node.val
                
            if node.val > low:
                subtree_sum += traverse(node.left, low, high)

            if node.val < high:
                subtree_sum +=  traverse(node.right, low, high)
                
            return subtree_sum
        
        return traverse(root, low, high)
```

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class BSTIterator:

    # When analyzing amortized time complexities, I find it easiest to reason that each node gets pushed and popped exactly once in next() when # 
    # iterating over all N nodes.
    # That comes out to 2N * O(1) over N calls to next(), making it O(1) on average, or O(1) amortized.

    def __init__(self, root: Optional[TreeNode]):
        
        self.stack = []
        
        # init minimum non-existent node
        node = root
        
        while node:
            self.stack.append(node)
            node = node.left            

    def next(self) -> int:
        
        popped = self.stack.pop()
        
        node = popped.right
                    
        while node:    
            self.stack.append(node)
            node = node.left
            
        return popped.val

    def hasNext(self) -> bool:
        
        if self.stack:
            return True
        
        else:
            return False
        


# Your BSTIterator object will be instantiated and called as such:
# obj = BSTIterator(root)
# param_1 = obj.next()
# param_2 = obj.hasNext()
```

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    
    diameter = float('-inf')
    
    def get_height_update_diameter(self, root):
                
        if not root:
            return 0
        
        left_height = self.get_height_update_diameter(root.left)
        right_height = self.get_height_update_diameter(root.right)
        
        dia = left_height + right_height
        
        if self.diameter < dia:
            self.diameter = dia
        
        return 1 + max(left_height, right_height)
        
    def diameterOfBinaryTree(self, root: Optional[TreeNode]) -> int:
        
        self.get_height_update_diameter(root)
        
        return self.diameter
```

```python
from collections import defaultdict

class Graph:

    def __init__(self, v):
        self.adj = defaultdict(list)
        self.v = v
    
    def add_edge(self, u, v):
        self.adj[u].append(v)
        self.adj[v].append(u)

    def is_bipartite_helper(self, queue, colors):

        while queue:
            x = queue.pop(0)

            for neib in self.adj[x]:

                if neib == x:
                    # self loop
                    return False

                if colors[neib] == 0:
                    colors[neib] = colors[x] * -1 # invert colors
                    queue.append(neib)

                elif colors[neib] !=  colors[x] * -1:
                    return False

        return True

    def is_bipartite(self):

        colors = [0] * self.v

        queue = [0]
        colors[0] = 1

        for i in range(self.v):
            if colors[i] != 0:
                if not self.is_bipartite_helper(queue, colors):
                    return False

        return True

dag = Graph(5)
dag.add_edge(0, 1)
dag.add_edge(0, 2)
dag.add_edge(2, 3)
dag.add_edge(3, 1)
print(dag.is_bipartite())

# dag = Graph(5)
# dag.add_edge(0, 1)
# dag.add_edge(0, 2)
# dag.add_edge(2, 4)
# dag.add_edge(3, 1)
# dag.add_edge(3, 4)
# print(dag.is_bipartite())
```

```python
"""
# Definition for a Node.
class Node:
    def __init__(self, x: int, next: 'Node' = None, random: 'Node' = None):
        self.val = int(x)
        self.next = next
        self.random = random
"""

class Solution:

    def __init__(self):
        self.hashmap = {}
    
    # this is beautiful
    def copyRandomList_dfs(self, head):

        if head == None:
            return None

        if head in self.hashmap:
            return self.hashmap[head]

        node = Node(head.val)
        self.hashmap[head] = node

        # Recursively copy the remaining linked list starting once from the next pointer and then from the random pointer.
        # Thus we have two independent recursive calls.
        # Finally we update the next and random pointers for the new node created.
        node.next = self.copyRandomList(head.next)
        node.random = self.copyRandomList(head.random)

        return node
    
    def copyRandomList(self, head: 'Optional[Node]') -> 'Optional[Node]':
        
        node = head
        
        # dummy
        dummy = Node(0)
        prev = dummy
        
        hashmap = {}
        
        while node:
            if node not in hashmap:
                node_copy = Node(node.val)
                hashmap[node] = node_copy
            
            node_copy = hashmap[node]
            prev.next = node_copy
            
            # random pointer
            if node.random:
                if node.random not in hashmap:
                    hashmap[node.random] = Node(node.random.val)

                node_copy.random = hashmap[node.random]    
            else:
                node_copy.random = None
                
            prev = node_copy
            node = node.next
        
        prev.next = None
        head = dummy.next
        
        return head
    
    # weaving of linked lists but modifies the parameter
    def copyRandomList(self, head):
        """
        :type head: Node
        :rtype: Node
        """
        if not head:
            return head

        # Creating a new weaved list of original and copied nodes.
        ptr = head
        while ptr:

            # Cloned node
            new_node = Node(ptr.val, None, None)

            # Inserting the cloned node just next to the original node.
            # If A->B->C is the original linked list,
            # Linked list after weaving cloned nodes would be A->A'->B->B'->C->C'
            new_node.next = ptr.next
            ptr.next = new_node
            ptr = new_node.next

        ptr = head

        # Now link the random pointers of the new nodes created.
        # Iterate the newly created list and use the original nodes random pointers,
        # to assign references to random pointers for cloned nodes.
        while ptr:
            ptr.next.random = ptr.random.next if ptr.random else None
            ptr = ptr.next.next

        # Unweave the linked list to get back the original linked list and the cloned list.
        # i.e. A->A'->B->B'->C->C' would be broken to A->B->C and A'->B'->C'
        ptr_old_list = head # A->B->C
        ptr_new_list = head.next # A'->B'->C'
        head_new = head.next
        while ptr_old_list:
            ptr_old_list.next = ptr_old_list.next.next
            ptr_new_list.next = ptr_new_list.next.next if ptr_new_list.next else None
            ptr_old_list = ptr_old_list.next
            ptr_new_list = ptr_new_list.next
        return head_new
    
    def copyRandomList_1(self, head: 'Optional[Node]') -> 'Optional[Node]':
        
        # doesnt work on duplicates
        node = head
        
        # dummy
        dummy = Node(0)
        prev = dummy
        
        hashmap = {}
        
        while node:
            
            if node.val not in hashmap:
                node_copy = Node(node.val)
                hashmap[node.val] = node_copy
            
            node_copy = hashmap[node.val]
            prev.next = node_copy
            
            # random pointer
            if node.random:
                rand_val = node.random.val
                if rand_val not in hashmap:
                    nxt_rand = Node(rand_val)
                    hashmap[rand_val] = nxt_rand

                node_copy.random = hashmap[rand_val]    
            else:
                node_copy.random = None
                
            prev = node_copy
            node = node.next
        
        prev.next = None
        head = dummy.next
        
        return head
```

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right

from collections import defaultdict
from queue import Queue

class Solution:
    
    def verticalOrder(self, root: Optional[TreeNode]) -> List[List[int]]:
        
        if not root:
            return 
        
        q = Queue()
        q.put((root, 0))
        res = defaultdict(list)
        min_column = 0
        max_column = 0
        
        while not q.empty():
            
            node, col = q.get()
            
            res[col].append(node.val)
            min_column = min(min_column, col)
            max_column = max(max_column, col)

            if node.left:
                q.put((node.left, col - 1))
            
            if node.right:
                q.put((node.right, col + 1))
                
        return [res[col] for col in range(min_column, max_column + 1)]
    
    def dfs(self, root: Optional[TreeNode]) -> List[List[int]]:
        
        # it is clearly a top to bottom (implying bfs)
        # to my surprise, there is actually a way via dfs as well.
        # this additinally requires sorting by row as well which
        # comes natural in bfs
        
        res = defaultdict(list)
        
        def vertical_order(node, column):
            
            if not node:
                return
            
            res[column].append(node.val)
            
            vertical_order(node.left, column - 1)
            vertical_order(node.right, column + 1)
            
        
        vertical_order(root, 0)
        return res.values()
```

```python
class Solution:
    def findBuildings(self, heights: List[int]) -> List[int]:
        
        # this is the most optimal but
        # just as a concept:
        # monotonic stack could be used 
        # but there's no point here.
        
        cur_max = heights[-1]
        ans = [len(heights) - 1]
        
        i = len(heights) - 2
        
        while i >= 0:
            if heights[i] > cur_max:
                ans.append(i)
                cur_max = heights[i]
                
            i -= 1
                
        return ans[::-1]
```
