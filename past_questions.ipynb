```python
class Solution:
    def productExceptSelf(self, nums: List[int]) -> List[int]:
        
        left = []
        #right = [0] * len(nums)
        
        for i in range(len(nums)):
            
            if i == 0:
                left.append(1)
                
            else:
                left.append(nums[i-1] * left[i-1])
                     
#             j = len(nums) - 1 - i
            
#             if i == 0:
#                 right[j] = 1
                
#             else:
#                 right[j] = nums[j+1] * right[j+1]
        right = 1
    
        for i in reversed(range(len(nums))):
            left[i] = left[i] * right
            right *= nums[i]
            
        return left
```

```python
class Solution:
    def validPalindrome(self, s: str) -> bool:
        
        # O(1) space complexity in the worst case
        # O(n) time complexity
    
        
        def is_valid(s, i, j, n_delete):
        
            while i < j:
                if s[i] != s[j]:
                    
                    if n_delete == 0:
                        n_delete += 1

                        if is_valid(s, i+1, j, n_delete):
                            return True
                        
                        return is_valid(s, i, j-1, n_delete)
                        
                    else:
                        return False
                
                i += 1
                j -= 1
            
            return True
    
        n_delete = 0
        i, j = 0, len(s) - 1

        
        return is_valid(s, i, j, n_delete)
        
        # solution 1: avoids recursion but is stupid

        def is_valid(s, delete_right=True):
            
            n_delete = 0
            i, j = 0, len(s) - 1
    
            while i < j:
                if s[i] != s[j]:
                    
                    if n_delete == 0:
                        
                        n_delete += 1
                        
                        if delete_right == True:
                            j -= 1
                            if s[i] != s[j]:
                                return False
                        else:
                            i += 1
                            if s[i] != s[j]:
                                return False
                        
                    else:
                        return False
                
                i += 1
                j -= 1
            
            return True
        
        if is_valid(s, delete_right=True):
            return True
        
        return is_valid(s, delete_right=False)
            
        
```
