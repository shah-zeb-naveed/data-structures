```python
# Given a string representing an arithmetic expression with only addition and multiplication operators,return the result of the calculation. For example, for "2*3+4", return 10

# 6 + 4 = 10
# 2 + 4 * 3
# 2, 12 -> 14

# 3 + 4 * 66

def eval(s):
    result = 0
    last = 0
    num = 0
    op = '+'

    for c in s:

        if c.digit():
            num = num * 10 + int(c) # 66
            continue

        elif op == '+':
            result += last # 3
            last = num # 4
        
        elif:
            last *= num 

        op = c # *
        num = 0 # 0

    if op == '+':
        result += num

    else:
        last *= num # 4 * 66
        result += last

    return result

"""   Given an integer array and an integer number k. Return the k-th largest element in the array.
    array = [5, -3, 9, 1]
* k = 0 => return: 9
* k = 1 => return: 5
* k = 3 => return: -3

# k starts with 0

# [-3, 1, 5, 9, 11, 2]

# k log (N)
# k = 2
# [9, -1, 3] O(K)
# N log K -> O(1)

"""

from heapq import *

def klargest(arr, k):

    if k > len(arr):
        raise Exception('Invalid K')

    # init
    heap = arr[:k + 1] # k = 2 -> 3rd largest -> [-3, 1, 5] 
    heapify(heap) # [-3, 1, 5]

    for x in arr[k + 1:]: # 2
        if x > heap[0]: # 2 > 5 -> No
            heappop(heap) # arr = [5, 9, 11]
            heappush(x) # arr = [5, 9, 11]

    # k largest elements in my heap -> top sitll points to min of them (LOL: this was probably unncessary. Just needed to return heap[0]
    for i in range(k - 1):
        heappop(heap)

    return heap[0]
```
