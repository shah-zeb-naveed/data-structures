```python

class QuerySystem:
  
  # bad solution. N is too high and sparse to initialize with boolean array.

  def __init__(self, N):    
    self.array = [False] * N
    self.cache = [False] * N

  def set(self, query):
    idx = query[1] - 1
    self.array[idx] = True
    self.cache[idx] = idx + 1
    
    i = idx - 1
    
    while i >= 0 and self.array[i] != True:
      self.cache[i] = idx + 1
      i -= 1
    
    return
    
  def get(self, query):
    return self.cache[query[1] - 1]
    
def answerQueries(queries, N):
  
  q_sys = QuerySystem(N)
  res = []
  
  for query in queries:
    if query[0] == 1:
      q_sys.set(query)
    else:
      val = q_sys.get(query)
      res.append(val)  
      
  return res
```

```python
from heapq import heappop, heappush


class Solution:
  def answerQueries(self, queries, N):
      # min heap
      h = []
      res = []

      for q in queries:
          # set
          if q[0] == 1:
              # no need to push if heap is full (all TRUE)
              if len(h) < N:
                  heappush(h, q[1])
          # get
          else:
              # no TRUE
              if not h:
                  res.append(-1)
              # all TRUE
              elif len(h) == N:
                  res.append(q[1])
              # get index < min TRUE index
              elif q[1] <= h[0]:
                  res.append(h[0])
              else:
                  backup = []
                  # keep popping until we find a min TRUE index, or empty heap
                  while h and q[1] > h[0]:
                      backup.append(heappop(h))

                  # if found min TRUE index
                  if h:
                      res.append(h[0])
                  # empty heap, meaning get index > any of TRUE index
                  else:
                      res.append(-1)
                  
                  # add popped TRUE back to heap
                  for b in backup:
                      heappush(h, b)

      return res
```
